A local language-learning app powered by **Gemma**. It turns what a learner reads or uploads into personalized flashcards, quizzes, and weak-word review,so vocabulary sticks because it comes from *their* material, not a fixed deck.

## The daily-life problem

Learning a language isn’t hard because of a lack of material, it’s hard because of **forgetting**. People look up words, feel like they know them, then lose them a week later because those words never return in a context that matters.

Generic flashcard apps make this worse: they drill vocabulary that has nothing to do with what someone is actually reading or trying to say.

**Gemma Vocab Coach** is built around one loop:

> **read / upload → Gemma makes flashcards → quiz → review weak words → repeat**

## Why Gemma is the core (not a bolt-on)

Gemma generates the learning content. Without it, the app has nothing useful to give the learner.

| Learner action | What Gemma does |
|---|---|
| Pick a topic or paste text | Generates personalized flashcards (word, meaning, example) |
| Upload a PDF / document | Extracts vocabulary from *that* document |
| Import an Excel vocab list | Fills missing meanings / fields |
| Ask about an uploaded doc | Answers using only that document’s content |
| Take a quiz | Writes quiz questions from the learner’s own saved words |
| Review | Helps surface and reinforce weak words until they’re learned |

Supported focus languages include English, Spanish, Vietnamese, French, and Chinese.

## How Gemma is called in the app

The production app uses the Google GenAI / Gemma path (API key in `.env`).  
Below is a **minimal illustration** of the same idea judges should understand: Gemma turns learner material into flashcards.

````python
# Illustrative prompt pattern used by Gemma Vocab Coach
# (In the real app this lives in services/gemma.py and streams into the UI.)

FLASHCARD_PROMPT = """
You are a language tutor powered by Gemma.
Create {n} vocabulary flashcards for a {level} learner of {language}.
Topic / source text:
---
{source}
---
Skip words the learner already knows: {known_words}
For each card return: word, meaning (in {native}), one short example sentence.
Focus on words that matter for THIS text — not a generic word list.
""".strip()

example = FLASHCARD_PROMPT.format(
    n=5,
    level="A2",
    language="Spanish",
    native="English",
    source="Hoy fui al mercado y compré pan fresco, tomates, y un poco de queso.",
    known_words="pan, queso",
)
print(example)
````

## Example learner journey (what judges would see in the demo)

1. **Onboard** — choose Spanish, level A2, goal “travel / reading.”
2. **Upload or pick a topic** — e.g. a short market dialogue or class PDF.
3. **Gemma generates flashcards** from that material; known words are skipped.
4. **Quiz** — questions written from *those* saved words.
5. **Weak review** — missed words resurface until learned.
6. **Dashboard** — simple progress on what’s sticking.

Optional paths in the same loop: Excel vocab import (Gemma fills gaps), Ask-the-document Q&A (answers grounded in the upload only).
